```python
import pandas as pd
import matplotlib.pyplot as plt
#Group 17
#we can do it 
#Student projects
# Sample mining production data (simulated for demonstration)
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Gold (tons)': [120, 110, 95, 90, 85],
    'Platinum (tons)': [140, 135, 130, 125, 120],
    'Coal (million tons)': [250, 240, 230, 220, 210],
    'Iron Ore (million tons)': [70, 75, 72, 78, 80]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Show the first few rows
print("Mining Production Data:")
print(df)

# -------------------------
# Plot 1: Line Plot of Mineral Production Over Time
# -------------------------
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Gold (tons)'], marker='o', label='Gold')
plt.plot(df['Year'], df['Platinum (tons)'], marker='s', label='Platinum')
plt.plot(df['Year'], df['Coal (million tons)'], marker='^', label='Coal')
plt.plot(df['Year'], df['Iron Ore (million tons)'], marker='d', label='Iron Ore')
plt.title('South African Mineral Production (2018-2022)')
plt.xlabel('Year')
plt.ylabel('Production')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------
# Plot 2: Bar Chart for Year 2022 Production
# -------------------------
production_2022 = df[df['Year'] == 2022].iloc[0, 1:]
production_2022.plot(kind='bar', color='skyblue')
plt.title('Mining Production in 2022')
plt.ylabel('Production')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------
# Plot 3: Pie Chart of Mineral Contribution in 2022
# -------------------------
plt.figure(figsize=(7, 7))
production_2022.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Mineral Contribution to Total Production in 2022')
plt.ylabel('')  # Remove y-label
plt.tight_layout()
plt.show()

# -------------------------
# Data Analysis: Yearly Growth/Decline Rates
# -------------------------
df['Gold Change (%)'] = df['Gold (tons)'].pct_change() * 100
df['Coal Change (%)'] = df['Coal (million tons)'].pct_change() * 100
print("\nYearly Percentage Change in Gold and Coal Production:")
print(df[['Year', 'Gold Change (%)', 'Coal Change (%)']].round(2))

# -------------------------
# Plot 4: Subplots - Gold vs Coal Production
# -------------------------
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(df['Year'], df['Gold (tons)'], color='gold', marker='o')
ax[0].set_title('Gold Production Over Time')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Tons')
ax[0].grid(True)

ax[1].plot(df['Year'], df['Coal (million tons)'], color='black', marker='o')
ax[1].set_title('Coal Production Over Time')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Million Tons')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
